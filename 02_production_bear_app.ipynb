```python
# Confirm the exported model exists
path = Path()
path.ls(file_exts='.pkl')
```

```python
# Load the inference learner from the file:
learn_inf = load_learner(path/'export.pkl')
```

```python
# A file upload widget
btn_upload = widgets.FileUpload()
```

```python
# Fetch the image
img = PILImage.create(btn_upload.data[-1])
```

```python
# Display the image
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: display(img.to_thumb(128,128))
```

```python
# Run inference on the image
pred,pred_idx,probs = learn_inf.predict(img)
```

```python
# Display the output of inference
lbl_pred = widgets.Label()
lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
```

```python
# A classify button:
btn_run = widgets.Button(description='Classify')
```

```python
# Click event handler for classify button
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
# Create a UI with a vertical box of these widgets

btn_upload = widgets.FileUpload()

VBox([
    widgets.Label('Select your bear!'),
    btn_upload,
    btn_run,
    out_pl,
    lbl_pred
])
```
